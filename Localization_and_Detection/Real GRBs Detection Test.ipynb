Here are some optimizations of the threshold and the window.  Looks like a window of 0.1 seconds is good.  A significance level of 4.5 is also best.  We get detections of 22 of the 29 GRBs this way.  I think that's good enough until a better trigger is developed.  If I lower the threshold I get some spurrious detections.  

I think the next step is to verify that all of the algorithms are working and then go from there.

```python
sig_levels = [3.5,4.0,4.5,5.0]
sig_stats = dict([(sig_level,dict())for sig_level in sig_levels])

for sig_level in sig_levels:
    for det in spacecraft.detectors:
        det.noise = False
        det.window = 1.0
        det.sig_level = sig_level
    stats = dict([(real_grb.locdb[27:-51],[0,[0]]) for real_grb in real_grbs])
    for real_grb in real_grbs:
        trig = spacecraft.throw_grb(real_grb)
        dets_trigged = np.array([det['counts'] for det in trig]) > 0
        stats[real_grb.locdb[27:-51]] = [np.sum(dets_trigged),np.array([det['T0'] for det in trig])[dets_trigged]]
        #print real_grb.locdb[27:-51], np.sum(dets_trigged),np.array([det['T0'] for det in trig])[dets_trigged]
    sig_stats[sig_level] = stats
```

```python
for sig_stat in sig_stats:
    print sig_stat,len(sig_stats[sig_stat]),
    detected = 0
    for stat in sig_stats[sig_stat].values():
        detected += stat[0]
    detected = detected / 5,
    average = 0.
    dev = 0.
    for stat in sig_stats[sig_stat].values():
        if len(stat[1]) > 0.:
            average += np.average(stat[1])
            dev += np.std(stat[1])
    print detected[0],(average/detected)[0],(dev/detected)[0]
```
