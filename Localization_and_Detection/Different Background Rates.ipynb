## Find the most likely position of each random GRB

This just loops through every GRB in the lookup table and compares it to every GRB in the random sample.  Goal is to find the best match. Also should look at the statistical bounds of this.  This takes a bit so beware large samples.

Note that the statistical error is in the 1st column of the counts arrays and I'm adding (in quadrature) the statistical error from the training sample and the random sample to find the total error.  Then, I'm just doing a simpler +/- analysis to see what happens to the location.

```python
norms = [calcNorms(counts,training_counts) for counts in real_counts]
```

```python
real_counts_err = [addErrors(counts,training_counts) for counts in real_counts]
```

```python
norms_err = [calcNormsWithError(counts,training_counts,real_counts_err[idx]) for idx,counts in enumerate(real_counts)]
```

Find the minimum distance of each GRB.

```python
loc_mins = [[ind_norm.argmin() for ind_norm in norm] for norm in norms]
```

```python
loc_mins_errm = [[norm.argmin() for norm in norm_err[0]] for norm_err in norms_err]
```

```python
loc_mins_errp = [[norm.argmin() for norm in norm_err[1]] for norm_err in norms_err]
```

Now, calculate the distance from the real GRB to the training one picked out from the distance measuremnt above.

```python
def distance(training_grbs,real_grbs,mins):
    return [eph.separation(grb.eph,training_grbs[mins[idx]].eph)*180./np.pi for idx,grb in enumerate(real_grbs)]
```

```python
errors = [distance(training_grbs,real_grbs,loc_min) for loc_min in loc_mins]
```

```python
errors_errm = [distance(training_grbs,real_grbs,loc_min_errm) for loc_min_errm in loc_mins_errm]
```

```python
errors_errp = [distance(training_grbs,real_grbs,loc_min_errp) for loc_min_errp in loc_mins_errp]
```

Plot and save the cumulative distribution of this error.

```python
stats = np.zeros((10,3))
```

```python
plt.figure(figsize=(10,5))
for idx in range(10):
    plt.subplot(2,5,idx+1)
    hist_data = plt.hist(errors[idx],bins=100,normed=1, histtype='step', cumulative=True)
    hist_data_errm = plt.hist(errors_errm[idx],bins=100,normed=1, histtype='step', cumulative=True)
    hist_data_errp = plt.hist(errors_errp[idx],bins=100,normed=1, histtype='step', cumulative=True)
    avg_stat = np.average([hist_data_errm[1][np.abs(hist_data_errm[0] - 0.68).argmin()],
                       hist_data_errp[1][np.abs(hist_data_errp[0] - 0.68).argmin()]])
    stats[idx] = [hist_data[1][np.abs(hist_data[0] - 0.68).argmin()],avg_stat,rates[idx]]
    plt.plot()
```

The 1-sigma error is around 68%.  Quick function to find the distance value that most closely matches 0.68.

```python
for stat in stats:
    print 'Sys/Stat Error: {:,.2f},{:,.2f} ({})'.format(stat[0],stat[1],stat[2])
```

```python
plt.figure(figsize=(10,8))
plt.plot(stats[:,2],stats[:,0],'o',label='Sys. Error')
plt.plot(stats[:,2],stats[:,1],'o',label='Stat. Error')
plt.xlabel('Background Rate [counts/sec]')
plt.ylabel('Error [degrees]')
plt.xlim((0,1100))
plt.legend(loc='lower right')
plt.savefig('Error vs. Background Rate.pdf', transparent = True)
plt.show()
```
