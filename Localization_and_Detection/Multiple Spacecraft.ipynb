```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
from BurstSim.GRB import *
from BurstSim.Detector import *
from BurstSim.Spacecraft import *
from BurstSim.Stats import calcNorms, addErrors, calcNormsWithError, distance, halfway, distance2
```

```python
pointings = {'01': ('60:0:0','10:0:0'),
            '02': ('120:0:0','15:0:0'),
            '03': ('180:0:0','20:0:0'),
            '04': ('240:0:0','10:0:0'),
            '05': ('300:0:0','15:0:0'),
            '06': ('360:0:0','20:0:0')}
```

```python
lons = ["{}:0:0".format(lon) for lon in np.linspace(10,40,4)]
```

```python
spacecrafts = [Spacecraft(pointings, lon = lon) for lon in lons]
```

```python
res = 250
rr,dd = np.meshgrid(np.linspace(0,360,res,endpoint=False),np.linspace(-90,90,res))
exposure_positions = np.vstack([rr.ravel(),dd.ravel()])
```

```python
exposures = np.array([[[detector.exposure(position[0],position[1]) for position in exposure_positions.T] 
                      for detector in sc.detectors] for sc in spacecrafts])
```

```python
plt.figure(figsize=(20,5))
m = Basemap(projection='moll',lon_0=180,resolution='c')
m.drawparallels(np.arange(-90.,120.,30.))
m.drawmeridians(np.arange(0.,420.,60.))
x,y = m(rr,dd)
for sp in range(4):
    plt.subplot(1, 4, sp+1)
    m.pcolormesh(x,y,exposures[sp].sum(axis=0).reshape((res,res)))
plt.savefig('Four Exposures.png', transparent = True)
plt.show()
```

```python
tot_exposure = np.sum([exposure.sum(axis=0).reshape((res,res)) for exposure in exposures],axis=0)
```

```python
training_counts = [sc.throw_grbs(training_grbs,scaled=True) for sc in spacecrafts]
```

```python
real_positions = np.array(zip(360.*np.random.random_sample(2000),180.*np.random.random_sample(2000)-90.))
```

```python
exposures = np.array([[[detector.exposure(position[0],position[1]) for position in real_positions] 
                      for detector in sc.detectors] for sc in spacecrafts])
```

```python
mask = np.all(exposures.any(axis=1), axis=0)
```

```python
real_grbs = [GRB(position[0],position[1]) for position in real_positions[mask]]
```

```python
np.shape(real_grbs)
```

```python
pos = np.array([[grb.eph._ra*180./np.pi,grb.eph._dec*180./np.pi] for grb in real_grbs])
plt.figure(figsize=(8,10))
m = Basemap(projection='moll',lon_0=180,resolution='c')
m.drawparallels(np.arange(-90.,120.,30.))
m.drawmeridians(np.arange(0.,420.,60.))
x,y = m(pos[:,0],pos[:,1])
m.scatter(x,y,3,marker='o',color='k')
plt.show()
```

```python
real_counts = [sc.throw_grbs(real_grbs, scaled=True) for sc in spacecrafts]
```

```python
norms = [calcNorms(real_counts[idx], training_counts[idx]) for idx in np.arange(4)]
```

```python
real_counts_err = [addErrors(real_counts[idx],training_counts[idx]) for idx in np.arange(4)]
```

```python
norms_err = [calcNormsWithError(real_counts[idx],training_counts[idx],real_counts_err[idx]) for idx in np.arange(4)]
```

```python
loc_mins = [[ind_norm.argmin() for ind_norm in norm] for norm in norms]
```

```python
loc_mins_errm = [[norm.argmin() for norm in norm_err[0]] for norm_err in norms_err]
```

```python
loc_mins_errp = [[norm.argmin() for norm in norm_err[1]] for norm_err in norms_err]
```

```python
errors = [distance(training_grbs,real_grbs,loc_min) for loc_min in loc_mins]
```

```python
errors_errm = [distance(training_grbs,real_grbs,loc_min_errm) for loc_min_errm in loc_mins_errm]
```

```python
errors_errp = [distance(training_grbs,real_grbs,loc_min_errp) for loc_min_errp in loc_mins_errp]
```

```python
stats = np.zeros((4,3))
```

```python
plt.figure(figsize=(20,5))
for idx in range(4):
    plt.subplot(1,4,idx+1)
    hist_data = plt.hist(errors[idx],bins=100,normed=1, histtype='step', cumulative=True)
    hist_data_errm = plt.hist(errors_errm[idx],bins=100,normed=1, histtype='step', cumulative=True)
    hist_data_errp = plt.hist(errors_errp[idx],bins=100,normed=1, histtype='step', cumulative=True)
    avg_stat = np.average([hist_data_errm[1][np.abs(hist_data_errm[0] - 0.68).argmin()],
                       hist_data_errp[1][np.abs(hist_data_errp[0] - 0.68).argmin()]])
    stats[idx] = [hist_data[1][np.abs(hist_data[0] - 0.68).argmin()],avg_stat,np.linspace(10,40,4)[idx]]
    plt.plot()
```

```python
for stat in stats:
    print 'Sys/Stat Error: {:,.2f},{:,.2f} ({})'.format(stat[0],stat[1],stat[2])
```

```python
avg_positions = halfway(np.transpose([[training_grbs[loc] for loc in loc_min] for loc_min in loc_mins]))
```

```python
avg_positions_p = halfway(np.transpose([[training_grbs[loc] for loc in loc_min] for loc_min in loc_mins_errp]))
```

```python
avg_positions_m = halfway(np.transpose([[training_grbs[loc] for loc in loc_min] for loc_min in loc_mins_errm]))
```

```python
avg_errors = distance2(avg_positions,real_grbs)
avg_errors_p = distance2(avg_positions_p,real_grbs)
avg_errors_m = distance2(avg_positions_m,real_grbs)
```

```python
avg_data = plt.hist(avg_errors,bins=100,normed=1, histtype='step', cumulative=True)
avg_data_errm = plt.hist(avg_errors_m,bins=100,normed=1, histtype='step', cumulative=True)
avg_data_errp = plt.hist(avg_errors_p,bins=100,normed=1, histtype='step', cumulative=True)
```

```python
avg_stat = np.average([avg_data_errm[1][np.abs(avg_data_errm[0] - 0.68).argmin()],
                       avg_data_errp[1][np.abs(avg_data_errp[0] - 0.68).argmin()]])
```

```python
print 'Statistical Error: {:,.2f}'.format(avg_stat)
print 'Systematic Error: {:,.2f}'.format(avg_data[1][np.abs(avg_data[0] - 0.68).argmin()])
```
