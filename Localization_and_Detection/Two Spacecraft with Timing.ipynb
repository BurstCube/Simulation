```python
loc_mins = [[ind_norm.argmin() for ind_norm in norm] for norm in norms]
```

```python
stats = np.zeros((4,3))
```

```python
plt.figure(figsize=(20,5))
for idx in range(2):
    plt.subplot(1,2,idx+1)
    hist_data = plt.hist(errors[idx],bins=100,normed=1, histtype='step', cumulative=True)
    hist_data_errm = plt.hist(errors_errm[idx],bins=100,normed=1, histtype='step', cumulative=True)
    hist_data_errp = plt.hist(errors_errp[idx],bins=100,normed=1, histtype='step', cumulative=True)
    avg_stat = np.average([hist_data_errm[1][np.abs(hist_data_errm[0] - 0.68).argmin()],
                       hist_data_errp[1][np.abs(hist_data_errp[0] - 0.68).argmin()]])
    stats[idx] = [hist_data[1][np.abs(hist_data[0] - 0.68).argmin()],avg_stat,np.linspace(10,40,4)[idx]]
    plt.plot()
```

```python
res = 250
rr,dd = np.meshgrid(np.linspace(0,360,res,endpoint=False),np.linspace(-90,90,res))
training_positions_hires = np.vstack([rr.ravel(),dd.ravel()])
```

```python
exposures = np.array([[[detector.exposure(position[0],position[1]) for position in training_positions_hires.T] 
                      for detector in sc.detectors] for sc in spacecrafts])
```

```python
mask_hires = np.all(exposures.any(axis=1), axis=0)
```

```python
training_grbs = [GRB(position[0],position[1]) for position in training_positions_hires.T]
```

```python
det1 = spacecrafts[0].detectors[0]
det2 = spacecrafts[1].detectors[0]

alpha1,delta1 = det1.obs.radec_of(0,90)
s1 = np.array([np.cos(delta1)*np.cos(alpha1), np.cos(delta1)*np.sin(alpha1),np.sin(delta1)])
r1 = det1.obs.elev

alpha2,delta2 = det2.obs.radec_of(0,90)
s2 = np.array([np.cos(delta2)*np.cos(alpha2), np.cos(delta2)*np.sin(alpha2),np.sin(delta2)])
r2 = det2.obs.elev

S = r1*s1 - r2*s2
S_u = S/np.linalg.norm(S)
```

```python
Gs = [np.array([np.cos(grb.eph.a_dec)*np.cos(grb.eph.a_ra),
              np.cos(grb.eph.a_dec)*np.sin(grb.eph.a_ra),
              np.sin(grb.eph.a_dec)]) for grb in training_grbs]
```

```python
Bs = [np.dot(G_u,G_u - np.dot(G_u,S_u)*S_u)/np.linalg.norm(G_u - np.dot(G_u,S_u)*S_u) for G_u in Gs]
```

```python
deltas = np.sin(np.arccos(Bs))*np.linalg.norm(S)/eph.c
```

```python
deltas_ma = np.ma.array(deltas,mask = ~mask_hires)
```

```python
plt.figure(figsize=(16,20))
m = Basemap(projection='moll',lon_0=180,resolution='c')
m.drawparallels(np.arange(-90.,120.,30.))
m.drawmeridians(np.arange(0.,420.,60.))
x,y = m(rr,dd)
m.pcolormesh(x,y,10000*deltas_ma.reshape((res,res)))
m.colorbar()
#m.colorbar()
#plt.savefig('Single Exposure.png', transparent = True)
plt.show()
```
