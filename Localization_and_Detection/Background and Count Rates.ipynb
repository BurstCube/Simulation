```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import matplotlib as mpl
```

```python
mpl.rcParams.update({'font.size': 22})
```

```python
import pickle
import numpy as np
```

```python
bkgrnd_pkl_file = open('DiffBackgroundRates.pkl', 'rb')
bkgrnd = pickle.load(bkgrnd_pkl_file)
bkgrnd_pkl_file.close()
```

```python
counts_pkl_file = open('DiffCountRates.pkl', 'rb')
counts = pickle.load(counts_pkl_file)
counts_pkl_file.close()
```

```python
fig = plt.figure(figsize=(10,8))

ax1 = fig.add_subplot(111)
ax2 = ax1.twiny()

plt.ylim(1,10)

ax1.plot(bkgrnd[:,2],bkgrnd[:,0],'o')
ax1.plot(bkgrnd[:,2],bkgrnd[:,1],'o')
ax1.set_xlabel('Background Rate [counts/sec]')
ax1.set_ylabel('Error [degrees]')
ax1.set_xlim((0,1100))

ax2.plot(counts[:,2],counts[:,0],'r*-',markersize=15)
ax2.plot(counts[:,2],counts[:,1],'y*-',markersize=15)
ax2.set_xlabel('GRB Rate [counts/sec]')
ax2.set_ylabel('Error [degrees]')
ax2.set_xlim((90,11000))


b1 = mpl.lines.Line2D([],[], color='blue', marker='o',
                          linestyle='none')
b2 = mpl.lines.Line2D([], [], color='green', marker='o',
                          linestyle='none')

c1 = mpl.lines.Line2D([],[], color='red', marker='*',
                          markersize=15,linestyle='none')
c2 = mpl.lines.Line2D([], [], color='yellow', marker='*',
                          markersize=15, linestyle='none')

plt.legend([b1,b2,c1,c2],('Bkgrnd Rate Sys. Error',
                          'Bkgrnd Rate Stat. Error',
                            'GRB Rate Sys. Error',
                          'GRB Rate Stat. Error'),loc='upper right')

plt.savefig('Error vs. Rates.pdf', transparent = True)
plt.show()
```
